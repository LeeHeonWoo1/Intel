### Data Refactor

1차 시도에서 데이터 수집 이후에 학습을 진행했을 때, 크게 진전되는 모습을 보이지 않았다. 아무래도 단변량 데이터이다 보니 그런 것 같기도 하고, 제목과 조회수 간의 상관관계가 그렇게 크지 않은 것 같다는 피드백도 있었다. 게시글의 본문을 추가로 수집하여 데이터를 수정한다.

```python
import pandas as pd

df_game = pd.read_csv("./data/renewal_dataset_game.csv")
df_game
```

```python
# 완료 와 같은 제목은 삭제한다.
df_game = df_game.loc[df_game['title'] != "완료"].reset_index(drop = True)

# 결측치 및 중복값 확인
df_game.info() # 본문 내용에 38개의 결측치 확인
df_game = df_game.dropna(subset=['content'], axis = 0)
```

```python
print(df_game['content'].duplicated().sum().sum()) # 49개의 중복값 확인
df_game = df_game.drop_duplicates(subset=['content'], keep='first').reset_index(drop=True)
df_game
```

```python
df = pd.read_csv("./data/renewal_dataset1.csv")
df
```

```python
# 게시판 명만 남아있는 부분들은 삭제한다.
df = df.loc[df['title'] != "자유게시판"].reset_index(drop = True)
df = df.loc[df['title'] != "가입인사"].reset_index(drop = True)
df
```

```python
df.info() # 19건의 결측치 확인
```

```python
# 위에서 적용했던 방법 그대로 결측치 제거, 중복값 처리
df = df.dropna(subset=['content'], axis = 0)
print(df['content'].duplicated().sum()) # 49개의 중복값 확인
df = df.drop_duplicates(subset=['content'], keep='first').reset_index(drop=True)
df
```

```python
df_ssy = pd.DataFrame(pd.read_csv("./data/contents3.csv").values, columns=['title', 'content', 'views'])
df_ssy = df_ssy.dropna(subset=['content'], axis = 0)
df_ssy = df_ssy.drop_duplicates(subset=['content'], keep='first').reset_index(drop=True)
```

```python
df_ssy
```

```python
df_lhw = pd.DataFrame(pd.read_csv("./data/renewal_dataset.csv").values, columns=['title', 'content', 'views'])
df_lhw = df_lhw.dropna(subset=['content'], axis = 0)
df_lhw = df_lhw.drop_duplicates(subset=['content'], keep='first').reset_index(drop=True)
```

```python
df_lhw = df_lhw.loc[df_lhw['title'].str.find("견적") == -1].reset_index(drop=True)
df_lhw
```

```python
new_df = pd.concat([df, df_game, df_lhw, df_ssy]).reset_index(drop=True)
new_df
```

```python
new_df['views'] = new_df['views'].astype('string').str.replace(",", '')
```

```python
new_df['views'] = new_df['views'].apply(lambda x : str(float(x.replace("만", "")) * 10000) if "만" in x else x)
```

```python
new_df['views'] = new_df['views'].astype(float).astype(int)
```

```python
new_df = new_df.dropna(subset=['content'], axis = 0).reset_index(drop = True)
```

### 데이터 전처리
1차에서 진행했던 방법과 동일하게 제목, 본문 내용에 모두 적용한다.

```python
new_df.to_csv("./renew_dataset.csv", sep=",", index=False, encoding="utf8")
```
